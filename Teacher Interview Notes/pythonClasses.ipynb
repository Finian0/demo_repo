```python
import numpy as np
import matplotlib.pyplot as plt
# customer data
# For every customer we have the number of minutes they spent on our website over the course of a week

number_of_customers = 5
# data1 = np.random.choice
data2 = np.random.randint(0,30, (number_of_customers, 7))
```

```python
print(data2)
```

```python
class customerData:
    # First item we will set up in our class is an initialization function
    # the first input into a funciton defined inside of a class is "self" (acts as a placeholder for what we will
    # later on class a specific instance of our class)
    def __init__(self, dataset):
        self.data = dataset

    def plot(self, customer_index):
        row = self.data[customer_index]
        plt.plot(row)
        plt.show()
```

```python
Raising_Canes = customerData(data2)

print(type(Raising_Canes))

print(Raising_Canes.data)

Raising_Canes.plot(0)
```
