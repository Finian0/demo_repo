````python
import random 

def knapsack(capacity):
    """
    :param capcacity: integer representing the size of the knapsack
    :return: variable knapsack which represents index of items in items[] that were put into the knapsack
    """
    global items
    knapsack = []
    remainingCapacity = capacity
    terminateFlag = 0
    for item in items:
        if remainingCapacity == 0: break
        if item[0] <= remainingCapacity:
            knapsack.append(items.index(item))
            remainingCapacity -= item[0]
    for generation in range(50):
        bestOffspring = []
        knapsackValue = getValue(knapsack)
        for offspring in range(50):
            mutatedKnapsack = random.sample(knapsack, len(knapsack) - 1)
            remainingCapacity = capacity - getCapacity(mutatedKnapsack)
            randomItems = items.copy()
            random.shuffle(randomItems)
            for item in randomItems:
                if remainingCapacity == 0: break
                if remainingCapacity >= item[0] and items.index(item) not in mutatedKnapsack:
                    mutatedKnapsack.append(items.index(item))
                    remainingCapacity -= item[0]
            if getValue(mutatedKnapsack) > getValue(bestOffspring) and getValue(mutatedKnapsack) > knapsackValue:
                bestOffspring = mutatedKnapsack.copy()
        if getValue(bestOffspring) > knapsackValue:
            knapsack = bestOffspring.copy()
        if not bestOffspring:
            terminateFlag += 1
        else:
            terminateFlag = 0
        if terminateFlag == 2: break
    return knapsack


def getCapacity(knapsack):
    """
    :param knapsack: list of integers that reference the index of items[]
    :return: sum of weight of each item
    """
    global items
    knapsackWeight = 0
    for item in knapsack:
        knapsackWeight += items[item][0]
    return knapsackWeight

def getValue(knapsack):
    """
    :param knapsack: list of integers that reference the index of items[]
    :return: sum of value of each item
    """
    global items
    knapsackValue = 0
    for item in knapsack:
        knapsackValue += items[item][1]
    return knapsackValue
````
