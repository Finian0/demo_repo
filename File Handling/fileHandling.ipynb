```python
def histogram(word):
    wordDict = dict()
    for letter in word:
        if letter in wordDict.keys():
            wordDict[letter] += 1
        else:
            wordDict[letter] = 1
    return wordDict
```

```python
def anagrams(input_string):
    """
    :param input_string: String to search for all anagrams with
    :return: No return but writes all anagrams to output file
    """
    inputDict = histogram(input_string)
    with open('words.txt', 'r') as read_file:
        with open('output.txt', 'w') as write_file:
            for line in read_file:
                line = line.strip()
                lineDict = histogram(line)
                if inputDict == lineDict:
                    write_file.write(line + '\n')
```

```python
#anagrams('pots')
anagrams('resent')
```

```python
def heteropalindromes():
    wordSet = set()
    hpSet = set()
    with open('words.txt', 'r') as read_file:
        for line in read_file:
            line = line.strip()
            wordSet.add(line)
        for word in wordSet:
            if word[::-1] in wordSet and not((word[::-1], word) in hpSet) and word != word[::-1]:
                hpSet.add((word, word[::-1]))
    return hpSet
```
